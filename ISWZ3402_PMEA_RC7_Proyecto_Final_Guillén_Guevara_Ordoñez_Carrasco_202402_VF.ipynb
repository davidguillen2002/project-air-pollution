### II.I. Series Temporales

El análisis de **series temporales** se enfoca en estudiar datos ordenados cronológicamente para identificar patrones, tendencias, estacionalidades y relaciones en el tiempo. Este tipo de análisis es crucial para la predicción de fenómenos como el clima, la demanda energética y la calidad del aire.

- **Modelos ARIMA (AutoRegressive Integrated Moving Average):** Combinan términos autorregresivos (AR), integración (I) y promedio móvil (MA), siendo útiles para series estacionarias.
- **Modelos SARIMA:** Extienden ARIMA incorporando estacionalidad.
- **Redes LSTM (Long Short-Term Memory):** Redes neuronales recurrentes diseñadas para capturar dependencias temporales a largo plazo.

## III.IX. Consolidación y Procesamiento de Datos

En esta sección se detalla el proceso técnico realizado para consolidar, limpiar y procesar los datos utilizados en este proyecto. La metodología implementada garantiza que el conjunto de datos sea apto para modelado y análisis avanzados.

---

### III.IX.I. Carga y Exploración de Datos

1. **Importación de Archivos:**
   - Se procesaron 13 archivos que contienen datos meteorológicos y químicos, incluyendo variables como **PM2.5**, **CO**, **NO2**, y **O3**.
   - Se utilizó una función `load_data` que maneja excepciones para garantizar la carga segura de archivos y generar registros de éxito o error mediante **logging**.

2. **Unificación de Archivos:**
   - Los datos de estaciones meteorológicas y contaminantes se fusionaron mediante un índice temporal común (`Date_time`).
   - Se utilizó una estrategia de unión "outer join" para evitar la pérdida de datos y preservar la máxima cantidad de información posible.

---

### III.IX.II. Preprocesamiento de Datos

1. **Normalización Temporal:**
   - Se transformó la columna `Date_time` a un formato estándar, eliminando inconsistencias como valores terminados en `.0`.
   - Se eliminaron registros con valores nulos en `Date_time` y se configuró como índice para facilitar la manipulación temporal.

2. **Optimización de Memoria:**
   - Se redujo el consumo de memoria cambiando los tipos de datos numéricos (`float64`, `int64`) a tipos más eficientes (`float32`, `int32`), asegurando un rendimiento computacional adecuado para el análisis.

3. **Imputación de Valores Faltantes:**
   - Se empleó **SimpleImputer** con la estrategia de media para rellenar valores faltantes en las columnas numéricas.
   - Esta técnica asegura la completitud del dataset mientras minimiza el impacto en las distribuciones originales.

---

### III.IX.III. Visualización de Datos Faltantes

1. **Análisis de Datos Faltantes:**
   - Se utilizó la biblioteca **MissingNo** para visualizar los patrones de datos faltantes a través de un **mapa de matriz** y un **mapa de calor**.
   - Estas visualizaciones ayudaron a identificar variables con grandes proporciones de valores ausentes y patrones de correlación entre columnas faltantes.

---

### III.IX.IV. Generación de Métricas Derivadas

1. **Creación de la Métrica `PM_Avg`:**
   - Se calculó el promedio de partículas (PM2.5, PM10, etc.) a través de todas las estaciones meteorológicas, creando una columna derivada (`PM_Avg`) que captura una vista integrada de los niveles de partículas.

---

### III.IX.V. Almacenamiento y Resumen del Dataset Final

1. **Exportación del Dataset Procesado:**
   - El dataset final, denominado `processed_data_imputed.csv`, incluye todas las variables originales, imputaciones realizadas y métricas derivadas.

2. **Resumen del Dataset:**
   - Se generaron estadísticas descriptivas del dataset final para asegurar su completitud y validez antes de proceder al modelado.

---

```python
import pandas as pd
import numpy as np
import logging
import os
import missingno as msno
from sklearn.impute import SimpleImputer

# Configuración de logs
logging.basicConfig(level=logging.INFO, format="%(asctime)s - %(levelname)s - %(message)s")

# Lista de archivos de datos
files = [
    "CO.csv", "PM25.csv", "DIR.csv", "HUM.csv", "LLU.csv", "NO2.csv", "O3.csv",
    "PM10.csv", "PRE.csv", "RS.csv", "SO2.csv", "TMP.csv", "VEL.csv"
]
stations_file = "estaciones.csv"

def load_data(file):
    """Carga un archivo CSV y maneja excepciones."""
    try:
        data = pd.read_csv(file)
        logging.info(f"Data loaded successfully from {file} with shape {data.shape}")
        return data
    except Exception as e:
        logging.error(f"Error loading data from {file}: {e}")
        return None

def preprocess_data(df):
    """Preprocesa un dataset para asegurar índices únicos y consistencia."""
    try:
        # Convertir Date_time a datetime, manejando inconsistencias (.0)
        df["Date_time"] = pd.to_datetime(df["Date_time"].str.replace(".0", "", regex=False), errors="coerce")
        # Eliminar filas con Date_time nulo
        df.dropna(subset=["Date_time"], inplace=True)
        # Configurar Date_time como índice
        df.set_index("Date_time", inplace=True)
        logging.info("Preprocessed data successfully")
    except Exception as e:
        logging.error(f"Error preprocessing data: {e}")
    return df

def optimize_memory(df):
    """Optimiza los tipos de datos en un DataFrame para reducir el uso de memoria."""
    try:
        for col in df.select_dtypes(include=["float64", "int64"]).columns:
            df[col] = pd.to_numeric(df[col], downcast="float")
        logging.info("Memory optimization completed")
    except Exception as e:
        logging.error(f"Error optimizing memory: {e}")
    return df

def visualize_missing_data(df):
    """Visualiza los datos faltantes usando MissingNo."""
    try:
        msno.matrix(df, figsize=(12, 6), fontsize=10)
        msno.heatmap(df, figsize=(12, 6))
        logging.info("Missing data visualization completed")
    except Exception as e:
        logging.error(f"Error visualizing missing data: {e}")

# Cargar datasets
datasets = {}
for file in files:
    datasets[file.split(".")[0]] = load_data(file)

# Cargar estaciones
stations_data = None
if os.path.exists(stations_file):
    try:
        stations_data = pd.read_csv(stations_file)
        logging.info(f"Stations data loaded successfully with shape {stations_data.shape}")
    except Exception as e:
        logging.error(f"Error loading data from {stations_file}: {e}")

# Preprocesar y fusionar datasets
merged_data = None
for name, data in datasets.items():
    if data is not None:
        # Preprocesar cada dataset
        data = preprocess_data(data)
        # Optimizar memoria
        data = optimize_memory(data)
        # Fusionar manejando conflictos de sufijos
        if merged_data is None:
            merged_data = data
        else:
            merged_data = pd.merge(
                merged_data, 
                data, 
                how="outer", 
                left_index=True, 
                right_index=True, 
                suffixes=("", f"_{name}")
            )

logging.info(f"Merged dataset shape: {merged_data.shape}")

# Visualizar datos faltantes
visualize_missing_data(merged_data)

# Imputar datos faltantes usando imputación básica (media)
try:
    imputer = SimpleImputer(strategy="mean")
    imputed_data = pd.DataFrame(imputer.fit_transform(merged_data), columns=merged_data.columns, index=merged_data.index)
    logging.info("Missing values imputed using mean strategy")
except Exception as e:
    logging.error(f"Error imputing missing values: {e}")

# Guardar datos procesados
try:
    imputed_data.to_csv("processed_data_imputed.csv")
    logging.info("Processed data saved to processed_data_imputed.csv")
except Exception as e:
    logging.error(f"Error saving processed data: {e}")

# Crear columna promedio de partículas (PM)
try:
    pm_columns = [col for col in imputed_data.columns if "PM" in col.upper()]
    if pm_columns:
        imputed_data["PM_Avg"] = imputed_data[pm_columns].mean(axis=1)
        logging.info(f"Created PM_Avg column using {len(pm_columns)} PM-related columns")
    else:
        logging.warning("No PM-related columns found to calculate average")
except Exception as e:
    logging.error(f"Error creating PM_Avg column: {e}")

# Resumen de datos finales
try:
    logging.info(f"Final dataset summary:\n{imputed_data.describe(include='all')}")
except Exception as e:
    logging.error(f"Error generating dataset summary: {e}")
```

## III.X. Exploración de Datos y Análisis de Patrones

Esta sección describe las técnicas implementadas para explorar, validar y analizar patrones temporales y estacionales en el dataset procesado, así como el uso de técnicas de reducción de dimensionalidad para identificar relaciones clave entre las variables.

---

### III.X. Construcción de Modelos Predictivos

#### **Carga y Preprocesamiento del Conjunto de Datos**
1. **Carga de Datos**  
   El archivo `processed_data_imputed.csv` se cargó exitosamente, confirmando una estructura adecuada para el análisis, incluyendo un total de **118,503 filas y columnas correspondientes a variables meteorológicas y contaminantes**.

#### **Ingeniería de Características**
En esta etapa, se generaron nuevas variables para enriquecer el modelo predictivo:
1. **Lags (Retrasos):**  
   - Se crearon variables retrasadas (lags) con valores de 1, 7 y 30 días para capturar patrones temporales recientes y estacionales.
   
2. **Estadísticas Móviles:**  
   - Se calcularon promedios y desviaciones estándar en ventanas de 7 días para modelar tendencias a corto plazo.

3. **Características Temporales:**  
   - Variables como el día de la semana, mes, trimestre y año se agregaron para capturar patrones estacionales y temporales relevantes.

Estas características permiten a los modelos identificar tanto dependencias temporales como relaciones no lineales entre variables.

```python
import pandas as pd
import numpy as np
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
import logging

# Configuración de logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Funciones auxiliares
# Cargar datos
def load_data(file_path):
    try:
        data = pd.read_csv(file_path, parse_dates=["Date_time"])
        logging.info(f"Dataset '{file_path}' loaded successfully with shape {data.shape}")
        return data
    except Exception as e:
        logging.error(f"Error loading dataset '{file_path}': {e}")
        raise

# Preprocesar datos
def preprocess_data(data):
    logging.info("Starting data preprocessing...")
    data.set_index("Date_time", inplace=True)
    data = data.interpolate(method="time")
    logging.info("Missing data imputed using time-based interpolation.")
    return data

# Crear features
def feature_engineering(data, pm_columns):
    logging.info("Starting feature engineering...")

    # Crear lag features
    lags = [1, 7, 30]
    for lag in lags:
        for col in pm_columns:
            data[f"{col}_lag_{lag}"] = data[col].shift(lag)

    # Crear rolling features
    for col in pm_columns:
        data[f"{col}_rolling_mean_7"] = data[col].rolling(window=7).mean()
        data[f"{col}_rolling_std_7"] = data[col].rolling(window=7).std()

    # Crear features temporales
    data["dayofweek"] = data.index.dayofweek
    data["month"] = data.index.month
    data["quarter"] = data.index.quarter
    data["year"] = data.index.year

    logging.info("Feature engineering completed.")
    return data.dropna()

# Entrenar modelos
def train_models(X_train, y_train, X_test, y_test):
    results = {}

    # Modelos a evaluar
    models = {
        "XGBRegressor": XGBRegressor(n_estimators=500, learning_rate=0.01, max_depth=7, random_state=42),
        "RandomForestRegressor": RandomForestRegressor(n_estimators=100, random_state=42)
    }

    for name, model in models.items():
        logging.info(f"Training {name}...")
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        # Métricas
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        mae = mean_absolute_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        mape = mean_absolute_percentage_error(y_test, y_pred)

        results[name] = {
            "model": model,
            "y_pred": y_pred,
            "rmse": rmse,
            "mae": mae,
            "r2": r2,
            "mape": mape
        }

        logging.info(f"{name} - RMSE: {rmse}, MAE: {mae}, R2: {r2}, MAPE: {mape}")

    return results

# Graficar resultados
def plot_results(results, y_test, target_station):
    plt.figure(figsize=(14, 7))
    for name, result in results.items():
        plt.plot(y_test.index, result["y_pred"], label=f"{name} Prediction")

    # Valores reales
    plt.plot(y_test.index, y_test, label="Real", color="blue", linestyle="-", linewidth=1.5)

    # Limitar las fechas a partir de julio de 2017
    plt.xlim(pd.Timestamp("2017-07-01"), y_test.index.max())

    plt.title(f"Predicción de PM2.5 para {target_station} vs Valores Reales", fontsize=16)
    plt.xlabel("Fecha", fontsize=14)
    plt.ylabel("PM2.5", fontsize=14)
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()

# Script principal
if __name__ == "__main__":
    file_path = "processed_data_imputed.csv"
    data = load_data(file_path)

    # Columnas de PM2.5
    pm_columns = ["BELISARIO_PM25", "CARAPUNGO_PM25", "CENTRO_PM25", "COTOCOLLAO_PM25", "EL_CAMAL_PM25"]

    # Preprocesamiento
    data = preprocess_data(data)

    # Ingeniería de características
    data = feature_engineering(data, pm_columns)

    # Iterar sobre todas las estaciones
    for target_station in pm_columns:
        logging.info(f"Processing station: {target_station}")
        X = data.drop(columns=[target_station])
        y = data[target_station]

        # Dividir datos
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

        # Entrenar y evaluar
        results = train_models(X_train, y_train, X_test, y_test)

        # Graficar resultados
        plot_results(results, y_test, target_station)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from statsmodels.tsa.arima.model import ARIMA
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt
import logging

# Configuración de logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Función para cargar datos
def load_data(file_path, static_path):
    try:
        data = pd.read_csv(file_path, parse_dates=["Date_time"])
        stations = pd.read_csv(static_path)
        logging.info(f"Dataset '{file_path}' loaded successfully with shape {data.shape}")
        # Merge datos geográficos
        for station in stations["STATION"]:
            data[f"{station}_LAT"] = stations.loc[stations["STATION"] == station, "LAT"].values[0]
            data[f"{station}_LON"] = stations.loc[stations["STATION"] == station, "LON"].values[0]
        return data
    except Exception as e:
        logging.error(f"Error loading dataset '{file_path}': {e}")
        raise

# Preprocesar datos
def preprocess_data(data):
    logging.info("Starting data preprocessing...")
    data.set_index("Date_time", inplace=True)
    data = data.interpolate(method="time")
    logging.info("Missing data imputed using time-based interpolation.")
    return data

# Crear features
def feature_engineering(data, pm_columns):
    logging.info("Starting feature engineering...")
    lags = [1, 7, 30]
    for lag in lags:
        for col in pm_columns:
            data[f"{col}_lag_{lag}"] = data[col].shift(lag)

    for col in pm_columns:
        data[f"{col}_rolling_mean_7"] = data[col].rolling(window=7).mean()
        data[f"{col}_rolling_std_7"] = data[col].rolling(window=7).std()

    data["dayofweek"] = data.index.dayofweek
    data["month"] = data.index.month
    data["quarter"] = data.index.quarter
    data["year"] = data.index.year

    logging.info("Feature engineering completed.")
    return data.dropna()

# Entrenar modelos con TimeSeriesSplit
def train_models_with_tscv(X, y):
    results = {}
    models = {
        "XGBRegressor": XGBRegressor(n_estimators=500, learning_rate=0.01, max_depth=7, random_state=42),
        "RandomForestRegressor": RandomForestRegressor(n_estimators=100, random_state=42)
    }
    tscv = TimeSeriesSplit(n_splits=5)
    for name, model in models.items():
        rmse_scores, mae_scores, r2_scores, mape_scores = [], [], [], []
        for train_index, test_index in tscv.split(X):
            X_train, X_test = X.iloc[train_index], X.iloc[test_index]
            y_train, y_test = y.iloc[train_index], y.iloc[test_index]
            model.fit(X_train, y_train)
            y_pred = model.predict(X_test)
            rmse_scores.append(np.sqrt(mean_squared_error(y_test, y_pred)))
            mae_scores.append(mean_absolute_error(y_test, y_pred))
            r2_scores.append(r2_score(y_test, y_pred))
            mape_scores.append(mean_absolute_percentage_error(y_test, y_pred))

        results[name] = {
            "model": model,
            "rmse": np.mean(rmse_scores),
            "mae": np.mean(mae_scores),
            "r2": np.mean(r2_scores),
            "mape": np.mean(mape_scores)
        }
        logging.info(f"{name} - RMSE: {results[name]['rmse']}, MAE: {results[name]['mae']}, R2: {results[name]['r2']}, MAPE: {results[name]['mape']}")

    return results

# Implementar LSTM
def train_lstm(X_train, y_train, X_test, y_test):
    X_train = X_train.values.reshape((X_train.shape[0], X_train.shape[1], 1))
    X_test = X_test.values.reshape((X_test.shape[0], X_test.shape[1], 1))
    model = Sequential([
        LSTM(50, activation='relu', input_shape=(X_train.shape[1], 1)),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')
    model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=0)
    y_pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    mape = mean_absolute_percentage_error(y_test, y_pred)
    logging.info(f"LSTM - RMSE: {rmse}, MAE: {mae}, R2: {r2}, MAPE: {mape}")
    return model, y_pred

# Predicción a largo plazo
def forecast_long_term(model, X, steps=60):
    forecast = []
    current_input = X.iloc[-1].values.reshape(1, -1)
    for _ in range(steps):
        pred = model.predict(current_input)[0]
        forecast.append(pred)
        current_input = np.roll(current_input, -1)
        current_input[0, -1] = pred
    return forecast

# Resamplear datos
def resample_data(data, frequency="MS"):
    resampled = data.resample(frequency).mean()
    logging.info(f"Data resampled to {frequency} frequency.")
    return resampled

# Graficar resultados
def plot_results(results, y_test, long_term_forecast, target_station, freq="MS"):
    plt.figure(figsize=(14, 7))
    plt.plot(y_test.index, y_test, label="Real", color="blue", linestyle="-", linewidth=1.5)
    for name, result in results.items():
        y_pred = result["model"].predict(X.iloc[-len(y_test):])
        plt.plot(y_test.index, y_pred, label=f"{name} Prediction")
    future_dates = pd.date_range(start=y_test.index[-1], periods=len(long_term_forecast) + 1, freq=freq)[1:]
    plt.plot(future_dates, long_term_forecast, label="Long-Term Forecast", linestyle="--", color="red")
    plt.title(f"Prediction and Forecast for {target_station}", fontsize=16)
    plt.xlabel("Date", fontsize=14)
    plt.ylabel("PM2.5", fontsize=14)
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()

# Script principal
if __name__ == "__main__":
    file_path = "processed_data_imputed.csv"
    static_path = "estaciones.csv"
    data = load_data(file_path, static_path)
    pm_columns = ["BELISARIO_PM25", "CARAPUNGO_PM25", "CENTRO_PM25", "COTOCOLLAO_PM25", "EL_CAMAL_PM25"]
    data = preprocess_data(data)
    data = feature_engineering(data, pm_columns)

    for target_station in pm_columns:
        logging.info(f"Processing station: {target_station}")
        X = data.drop(columns=[target_station])
        y = data[target_station]
        results = train_models_with_tscv(X, y)

        # Predicción a largo plazo (mensual)
        best_model = max(results.items(), key=lambda x: x[1]["r2"])[1]["model"]
        long_term_forecast = forecast_long_term(best_model, X, steps=60)

        # Resamplear datos (mensual)
        resampled_data = resample_data(data, frequency="MS")

        plot_results(results, y.iloc[-len(long_term_forecast):], long_term_forecast, target_station, freq="MS")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from prophet import Prophet  # Actualización del nombre correcto del paquete
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt
import logging
import streamlit as st

# Configuración de logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# ==========================
# Función para cargar datos
# ==========================
def load_data(file_path, static_path):
    try:
        data = pd.read_csv(file_path, parse_dates=["Date_time"])
        stations = pd.read_csv(static_path)
        logging.info(f"Dataset '{file_path}' loaded successfully with shape {data.shape}")
        
        # Merge datos geográficos
        for station in stations["STATION"]:
            data[f"{station}_LAT"] = stations.loc[stations["STATION"] == station, "LAT"].values[0]
            data[f"{station}_LON"] = stations.loc[stations["STATION"] == station, "LON"].values[0]
        return data
    except Exception as e:
        logging.error(f"Error loading dataset '{file_path}': {e}")
        raise

# ==========================
# Preprocesar datos
# ==========================
def preprocess_data(data):
    logging.info("Starting data preprocessing...")
    data.set_index("Date_time", inplace=True)
    data = data.interpolate(method="time")
    logging.info("Missing data imputed using time-based interpolation.")
    return data

# ==========================
# Crear features
# ==========================
def feature_engineering(data, pm_columns):
    logging.info("Starting feature engineering...")
    lags = [1, 7, 30]
    for lag in lags:
        for col in pm_columns:
            data[f"{col}_lag_{lag}"] = data[col].shift(lag)

    for col in pm_columns:
        data[f"{col}_rolling_mean_7"] = data[col].rolling(window=7).mean()
        data[f"{col}_rolling_std_7"] = data[col].rolling(window=7).std()

    data["dayofweek"] = data.index.dayofweek
    data["month"] = data.index.month
    data["quarter"] = data.index.quarter
    data["year"] = data.index.year

    logging.info("Feature engineering completed.")
    return data.dropna()

# ==========================
# Entrenar modelos con TSCV
# ==========================
def train_models_with_tscv(X, y):
    results = {}
    models = {
        "XGBRegressor": XGBRegressor(n_estimators=500, learning_rate=0.01, max_depth=7, random_state=42),
        "RandomForestRegressor": RandomForestRegressor(n_estimators=100, random_state=42)
    }
    tscv = TimeSeriesSplit(n_splits=5)
    for name, model in models.items():
        param_grid = {
            'n_estimators': [100, 200, 300],
            'max_depth': [5, 10, 15]
        } if name == "XGBRegressor" else {
            'n_estimators': [100, 200, 300],
            'max_depth': [10, 20, 30]
        }

        grid_search = GridSearchCV(model, param_grid, cv=tscv, scoring='neg_mean_squared_error')
        grid_search.fit(X, y)
        best_model = grid_search.best_estimator_
        logging.info(f"{name} - Best Params: {grid_search.best_params_}")

        y_pred = best_model.predict(X)
        results[name] = {
            "model": best_model,
            "rmse": np.sqrt(mean_squared_error(y, y_pred)),
            "mae": mean_absolute_error(y, y_pred),
            "r2": r2_score(y, y_pred)
        }
        logging.info(f"{name} - RMSE: {results[name]['rmse']}, MAE: {results[name]['mae']}, R2: {results[name]['r2']}")

    return results

# ==========================
# Implementar Prophet
# ==========================
def run_prophet(data, station):
    prophet_data = data[["Date_time", station]].rename(columns={"Date_time": "ds", station: "y"})
    model = Prophet()
    model.fit(prophet_data)
    
    future = model.make_future_dataframe(periods=365)
    forecast = model.predict(future)

    return forecast, model

# ==========================
# Graficar resultados
# ==========================
def plot_results(results, y_test, target_station):
    plt.figure(figsize=(14, 7))
    for name, result in results.items():
        y_pred = result["model"].predict(y_test)
        plt.plot(y_test.index, y_pred, label=f"{name} Prediction")

    plt.title(f"Prediction for {target_station}")
    plt.xlabel("Date")
    plt.ylabel("PM2.5")
    plt.legend()
    plt.show()

# ==========================
# Streamlit App
# ==========================
def build_streamlit_app():
    st.title("Air Pollution Forecasting")

    data = load_data("processed_data_imputed.csv", "estaciones.csv")
    pm_columns = ["BELISARIO_PM25", "CARAPUNGO_PM25", "CENTRO_PM25", "COTOCOLLAO_PM25", "EL_CAMAL_PM25"]
    data = preprocess_data(data)
    data = feature_engineering(data, pm_columns)

    station = st.selectbox("Select a station:", pm_columns)

    st.write("### Model Training")
    X = data.drop(columns=pm_columns)
    y = data[station]
    results = train_models_with_tscv(X, y)
    st.write(results)

    st.write("### Prophet Forecast")
    forecast, _ = run_prophet(data, station)
    st.line_chart(forecast[["ds", "yhat"]].set_index("ds"))

if __name__ == "__main__":
    build_streamlit_app()
```
